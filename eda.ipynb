```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import preprocessing
```

```python
dta = load_iris()

data = pd.DataFrame(dta.data, columns=dta.feature_names)
```

```python
data.describe()
```

```python
for i in data.columns:
    sns.distplot(data[i].values)
    plt.title(f'{i}')
    plt.show()
```

```python
# lets preprocess the data n apply some transformations onto it
scale = preprocessing.MinMaxScaler(data)
```

```python
# since i saw bimodal characteristic in the data 
# i am reducing the 2 columns
petal_length_max = data['petal length (cm)'].max()
petal_width_max = data['petal width (cm)'].max()
data['petal length (cm)'] = data['petal length (cm)']/petal_length_max
data['petal width (cm)'] = data['petal width (cm)']/petal_width_max
```

```python
for i in data.columns:
    sns.distplot(data[i].values)
    plt.title(f'{i}')
    plt.show()
```
